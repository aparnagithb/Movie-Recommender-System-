```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv', on_bad_lines='skip')
credits.shape
#movies.shape
```

```python
movies =movies.merge(credits, on ='title')
```

```python
movies['original_language'].value_counts()#most of en so can drop it
```

```python
movies=movies[['id', 'title', 'overview','genres','keywords', 'cast', 'crew']]
movies.head()
#we'll recommend movies with actor names , genre
```

```python
movies.dropna(inplace=True) #drops rows with null values
movies.iloc[0].genres
```

we want it like action , fanatsy , adventure

```python
import ast
def convert(obj):
  try :
    obj = ast.literal_eval(obj)
  except:
    return []

  l=[]
  for i in obj :
    l.append(i['name'])
  return l
convert(movies.iloc[0].genres)
#The error TypeError: string indices must be integers typically occurs when you try to treat a string as a dictionary or list , here the entry is string
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['genres']
```

```python
movies['keywords']
```

```python
import ast

def extract_names(obj):
    # Safely evaluate the string representation of a list of dictionaries
    items = ast.literal_eval(obj)

    # Extract 'name' from each dictionary and append to the list
    names_list = [item['name'] for item in items]

    return names_list

# Assuming 'keywords' is a column in the DataFrame
movies['keywords'] = movies['keywords'].apply(extract_names)
```

```python
movies['keywords']
```

```python
def convert3(obj): #just top 3 actors
  l=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter!=3:
      l.append(i['name'])
      counter+=1
    else:
      break
  return l
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
    movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

convert joe dohn to joedohn cuz it would be considered as two different persons another joe somewhere would be considered same as this joe

```python
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['crew']+movies['cast']
```

```python
newdf=movies[['id', 'title','tags']]
newdf['tags']=newdf['tags'].apply(lambda x:" ".join(x))
```

```python
newdf['tags'][0]
```

```python
newdf['tags'][1]
```

we want to find the similarity between the above two texts , but since they are words , we need to convert them to vector

```python
newdf['tags']=newdf['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors=cv.fit_transform(newdf['tags']).toarray() #convert from scipy sparse matrix to numpy arrray
vectors[0]
```

```python
#stemming
#[loved , loving , love]
#[love love love]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer ()
```

```python
ps.stem("loveable")
```

```python
ps.stem("loved")#you will get the root word , which has been mapped for a set of words
```

```python
ps.stem("lovely")
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
newdf['tags']=newdf['tags'].apply(stem)
```
